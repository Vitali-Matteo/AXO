```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import  make_moons
```

```python
X, y = make_moons(n_samples=300, noise=0.1)

fig = plt.figure()

plt.scatter(X[:, 0], X[:, 1], c = y)

n, p = X.shape
```

```python
def kernel_f(x, y, gamma = 3):
  return np.exp(-gamma * np.linalg.norm(x - y) ** 2)

K = np.array([[kernel_f(X[i], X[j]) for i in range(n)] for j in range(n)])

n = K.shape[0]
one_n = np.ones((n, n)) / n
K_bar = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)

U, S, VT = np.linalg.svd(K_bar, full_matrices=False)

Phi = U @ np.diag(S)

fig = plt.figure()
plt.scatter(Phi[:, 0], Phi[:, 1], c = y)
plt.axvline(x=0, color='r', linestyle='--')
```

```python
def kernel(x_i, x_j, q = 3):
  return (1 + x_i.T @ x_j) ** q
```

```python
#I want to find with least squares the bounding region of an ellipse made with random Gaussian points

a = 4 * 0.3
b = 2
n_points = 150

X = np.arange(-1, 2, 3 / n_points).reshape(-1, 1) #np.random.normal(0, a, (1000, 1))
Y = 3 * (X ** 3 ) - 2 * (X ** 2) + np.random.normal(0, b, (n_points, 1))

n, p = X.shape

plt.scatter(X, Y, marker="x")

K = np.array([[kernel(X[i], X[j]) for i in range(n)] for j in range(n)])

lam = 1.0
beta = np.linalg.inv(K + lam * np.eye(K.shape[0])) @ Y

y_pred = np.empty((n, 1))
for j in range(n):
  x_j = X[j]
  y_pred[j] = beta.T @ np.array([kernel(X[i], x_j) for i in range(n)])

plt.plot(X, y_pred, "r--")

y_true = 3 * (X ** 3 ) - 2 * (X ** 2)
plt.plot(X, y_true, "g--")
plt.legend(["Noisy data", "Predicted", "True"])
```
